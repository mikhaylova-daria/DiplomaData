```python
import csv
import sys
import os
import pandas
pathCorpus="/home/daria/Diploma/RandomCorpus"
```

```python
def transform(folder): 
    print folder
    articles = os.listdir(unicode(pathCorpus+os.sep+folder))
    for article in articles:
        print article
        f=open(unicode(pathCorpus+os.sep+folder+os.sep+article+os.sep+'features'))
        df = pandas.read_csv(filepath_or_buffer=f, sep='\t')
        if (len(df.index)>113):
            del (df['id'])
            del (df['id.1'])
            df.to_csv(path_or_buf=pathCorpus+os.sep+folder+os.sep+article+os.sep+"features", sep='\t', index_label="id")
```

```python
import multiprocessing 
pool = multiprocessing.Pool(processes=multiprocessing.cpu_count())
pool.map(transform, os.listdir(pathCorpus))
```
